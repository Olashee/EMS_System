```python
D = {101: {'name': 'Satya', 'age': 27, 'department': 'HR', 'salary': 50000}}
```

```python

D = {}

def add_employee():
  Employee=input("Enter Employee ID : ")
  if(Employee in D):
      print("Employee already exists")
  else:
    Name=input("Enter Employee Name : ")
    Age=int(input("Enter Employee Age : "))
    Department=input("Enter Employee Department : ")
    Salary=input("Enter Employee Salary : ")
    D[Employee]={'name':Name,'age':Age,'department':Department,'salary':Salary}
    print("Employee added successfully")

def view_employee():
   if(D):
    print("+------------+----------+-----+-------------+--------+")
    print("| EmployeeID | Name     | Age | Department  | Salary |")
    print("+------------+----------+-----+-------------+--------+")
    for emp_id, details in D.items():
      print(f"|{emp_id:<10} | {details['name']:<8} | {details['age']:<4} | {details['department']:<12} | {details['salary']:<8}|")
      print("+------------+----------+-----+-------------+--------+")
    else:
      print("No Employees available")

def search_employee():
  emp_id=input("Enter Employee ID : ")
  if(emp_id in D):
    print("Employee ID : ",emp_id)
    print("Employee Name : ",D[emp_id]['name'])
    print("Employee Age : ",D[emp_id]['age'])
    print("Employee Department : ",D[emp_id]['department'])
  elif emp_id not in D:
    print("Employee not found")

def Main_Menu(args):
  if args == '1':
    add_employee()
  elif args == '2':
    view_employee()
  elif args == '3':
    search_employee()
  elif args == '4':
    print("Thank you for using EMS")
    return True

  else:
    print("Invalid choice")
    return False



def main():
  print("Welcome to EMS")
  print("------------------------------------------------------------------")
  print("------------------------------------------------------------------")
  while True:
    value=input(
      "Enter your choice :"+"\n1.Add Employee"+"\n2.View All Employees"+
      "\n3.Search for Employee" + "\n4.Exit"+"\n")

    if Main_Menu(value):
      break


if __name__ == "__main__":
  main()
```
